## **4. Tokenize dataset**

### **4.1. Tokenize train set**

```python
# Äá»‹nh nghÄ©a hÃ m preprocess_training_examples vÃ  nháº­n Ä‘á»‘i sá»‘ examples lÃ  dá»¯ liá»‡u Ä‘Ã o táº¡o
def preprocess_training_examples(examples):
    # TrÃ­ch xuáº¥t danh sÃ¡ch cÃ¢u há»i tá»« examples vÃ  loáº¡i bá» cÃ¡c khoáº£ng tráº¯ng dÆ° thá»«a
    questions = [q.strip() for q in examples["question"]]

    # Tiáº¿n hÃ nh mÃ£ hÃ³a thÃ´ng tin Ä‘áº§u vÃ o sá»­ dá»¥ng tokenizer
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=MAX_LENGTH,
        truncation="only_second",
        stride=STRIDE,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    # TrÃ­ch xuáº¥t offset_mapping tá»« inputs vÃ  loáº¡i bá» nÃ³ ra khá»i inputs
    offset_mapping = inputs.pop("offset_mapping")

    # TrÃ­ch xuáº¥t sample_map tá»« inputs vÃ  loáº¡i bá» nÃ³ ra khá»i inputs
    sample_map = inputs.pop("overflow_to_sample_mapping")

    # TrÃ­ch xuáº¥t thÃ´ng tin vá» cÃ¢u tráº£ lá»i (answers) tá»« examples
    answers = examples["answers"]

    # Khá»Ÿi táº¡o danh sÃ¡ch cÃ¡c vá»‹ trÃ­ báº¯t Ä‘áº§u vÃ  káº¿t thÃºc cÃ¢u tráº£ lá»i
    start_positions = []
    end_positions = []

    # Duyá»‡t qua danh sÃ¡ch offset_mapping
    for i, offset in enumerate(offset_mapping):
        # XÃ¡c Ä‘á»‹nh index cá»§a máº«u (sample) liÃªn quan Ä‘áº¿n offset hiá»‡n táº¡i
        sample_idx = sample_map[i]

        # TrÃ­ch xuáº¥t sequence_ids tá»« inputs
        sequence_ids = inputs.sequence_ids(i)

        # XÃ¡c Ä‘á»‹nh vá»‹ trÃ­ báº¯t Ä‘áº§u vÃ  káº¿t thÃºc cá»§a ngá»¯ cáº£nh
        idx = 0
        while sequence_ids[idx] != 1:
            idx += 1
        context_start = idx
        while sequence_ids[idx] == 1:
            idx += 1
        context_end = idx - 1

        # TrÃ­ch xuáº¥t thÃ´ng tin vá» cÃ¢u tráº£ lá»i cho máº«u nÃ y
        answer = answers[sample_idx]

        if len(answer['text']) == 0:
            start_positions.append(0)
            end_positions.append(0)
        else:
            # XÃ¡c Ä‘á»‹nh vá»‹ trÃ­ kÃ½ tá»± báº¯t Ä‘áº§u vÃ  káº¿t thÃºc cá»§a cÃ¢u tráº£ lá»i trong ngá»¯ cáº£nh
            start_char = answer["answer_start"][0]
            end_char = answer["answer_start"][0] + len(answer["text"][0])

            # Náº¿u cÃ¢u tráº£ lá»i khÃ´ng náº±m hoÃ n toÃ n trong ngá»¯ cáº£nh, gÃ¡n nhÃ£n lÃ  (0, 0)
            if offset[context_start][0] > start_char or offset[context_end][1] < end_char:
                start_positions.append(0)
                end_positions.append(0)
            else:
                # Náº¿u khÃ´ng, gÃ¡n vá»‹ trÃ­ báº¯t Ä‘áº§u vÃ  káº¿t thÃºc dá»±a trÃªn vá»‹ trÃ­ cá»§a cÃ¡c mÃ£ thÃ´ng tin
                idx = context_start
                while idx <= context_end and offset[idx][0] <= start_char:
                    idx += 1
                start_positions.append(idx - 1)

                idx = context_end
                while idx >= context_start and offset[idx][1] >= end_char:
                    idx -= 1
                end_positions.append(idx + 1)

    # ThÃªm thÃ´ng tin vá»‹ trÃ­ báº¯t Ä‘áº§u vÃ  káº¿t thÃºc vÃ o inputs
    inputs["start_positions"] = start_positions
    inputs["end_positions"] = end_positions

    return inputs
```

```python
# Táº¡o má»™t biáº¿n train_dataset vÃ  gÃ¡n cho nÃ³ giÃ¡ trá»‹ sau khi Ã¡p dá»¥ng hÃ m preprocess_training_examples lÃªn táº­p dá»¯ liá»‡u "train"
# Báº­t cháº¿ Ä‘á»™ xá»­ lÃ½ theo tá»«ng batch báº±ng cÃ¡ch Ä‘áº·t batched=True
# Loáº¡i bá» cÃ¡c cá»™t khÃ´ng cáº§n thiáº¿t trong táº­p dá»¯ liá»‡u "train" báº±ng cÃ¡ch sá»­ dá»¥ng remove_columns

train_dataset = raw_datasets["train"].map(
    preprocess_training_examples,
    batched=True,
    remove_columns=raw_datasets["train"].column_names,
)

# In ra Ä‘á»™ dÃ i cá»§a táº­p dá»¯ liá»‡u "train" ban Ä‘áº§u vÃ  Ä‘á»™ dÃ i cá»§a táº­p dá»¯ liá»‡u Ä‘Ã£ Ä‘Æ°á»£c xá»­ lÃ½ (train_dataset)
len(raw_datasets["train"]), len(train_dataset)
```

### **4.2. Tokenize Val set**

```python
def preprocess_validation_examples(examples):
    # Chuáº©n bá»‹ danh sÃ¡ch cÃ¢u há»i báº±ng cÃ¡ch loáº¡i bá» khoáº£ng tráº¯ng á»Ÿ Ä‘áº§u vÃ  cuá»‘i má»—i cÃ¢u há»i
    questions = [q.strip() for q in examples["question"]]

    # Sá»­ dá»¥ng tokenizer Ä‘á»ƒ mÃ£ hÃ³a cÃ¡c cÃ¢u há»i vÃ  vÄƒn báº£n liÃªn quan
    inputs = tokenizer(
        questions,
        examples["context"],
        max_length=MAX_LENGTH,
        truncation="only_second",
        stride=STRIDE,
        return_overflowing_tokens=True,
        return_offsets_mapping=True,
        padding="max_length",
    )

    # Láº¥y Ã¡nh xáº¡ Ä‘á»ƒ Ã¡nh xáº¡ láº¡i vÃ­ dá»¥ tham chiáº¿u cho tá»«ng dÃ²ng trong inputs
    sample_map = inputs.pop("overflow_to_sample_mapping")
    example_ids = []

    # XÃ¡c Ä‘á»‹nh vÃ­ dá»¥ tham chiáº¿u cho má»—i dÃ²ng Ä‘áº§u vÃ o vÃ  Ä‘iá»u chá»‰nh Ã¡nh xáº¡ offset
    for i in range(len(inputs["input_ids"])):
        sample_idx = sample_map[i]
        example_ids.append(examples["id"][sample_idx])

        sequence_ids = inputs.sequence_ids(i)
        offset = inputs["offset_mapping"][i]

        # Loáº¡i bá» cÃ¡c offset khÃ´ng phÃ¹ há»£p vá»›i sequence_ids
        inputs["offset_mapping"][i] = [
            o if sequence_ids[k] == 1 else None for k, o in enumerate(offset)
        ]

    # ThÃªm thÃ´ng tin vÃ­ dá»¥ tham chiáº¿u vÃ o Ä‘áº§u vÃ o
    inputs["example_id"] = example_ids

    return inputs
```

```python
# Táº¡o má»™t biáº¿n validation_dataset vÃ  gÃ¡n giÃ¡ trá»‹ báº±ng viá»‡c sá»­ dá»¥ng dá»¯ liá»‡u tá»« raw_datasets["validation"] sau khi Ã¡p dá»¥ng má»™t hÃ m xá»­ lÃ½ trÆ°á»›c.
validation_dataset = raw_datasets["validation"].map(
    preprocess_validation_examples,  # Gá»i hÃ m preprocess_validation_examples Ä‘á»ƒ xá»­ lÃ½ dá»¯ liá»‡u Ä‘áº§u vÃ o.
    batched=True,  # Xá»­ lÃ½ dá»¯ liá»‡u theo tá»«ng batch.
    remove_columns=raw_datasets["validation"].column_names,  # Loáº¡i bá» cÃ¡c cá»™t khÃ´ng cáº§n thiáº¿t tá»« dá»¯ liá»‡u ban Ä‘áº§u.
)

# In ra Ä‘á»™ dÃ i cá»§a raw_datasets["validation"] vÃ  validation_dataset Ä‘á»ƒ so sÃ¡nh.
len(raw_datasets["validation"]), len(validation_dataset)
```

## **5. Train model**

```python
# Load model
model = AutoModelForQuestionAnswering.from_pretrained(MODEL_NAME)
```

```python
# Táº¡o Ä‘á»‘i tÆ°á»£ng args lÃ  cÃ¡c tham sá»‘ cho quÃ¡ trÃ¬nh huáº¥n luyá»‡n
args = TrainingArguments(
    output_dir="distilbert-finetuned-squadv2",  # ThÆ° má»¥c lÆ°u trá»¯ káº¿t quáº£ huáº¥n luyá»‡n
    evaluation_strategy="no",  # Cháº¿ Ä‘á»™ Ä‘Ã¡nh giÃ¡ khÃ´ng tá»± Ä‘á»™ng sau má»—i epoch
    save_strategy="epoch",  # LÆ°u checkpoint sau má»—i epoch
    learning_rate=2e-5,  # Tá»‘c Ä‘á»™ há»c
    num_train_epochs=3,  # Sá»‘ epoch huáº¥n luyá»‡n
    weight_decay=0.01,  # Giáº£m trá»ng lÆ°á»£ng mÃ´ hÃ¬nh Ä‘á»ƒ trÃ¡nh overfitting
    fp16=True,  # Sá»­ dá»¥ng kiá»ƒu dá»¯ liá»‡u half-precision (16-bit) Ä‘á»ƒ tá»‘i Æ°u hÃ³a tÃ i nguyÃªn
    push_to_hub=True,  # Äáº©y káº¿t quáº£ huáº¥n luyá»‡n lÃªn má»™t nÆ¡i chia sáº» trá»±c tuyáº¿n (Hub)
)
```

```python
# Khá»Ÿi táº¡o má»™t Ä‘á»‘i tÆ°á»£ng Trainer Ä‘á»ƒ huáº¥n luyá»‡n mÃ´ hÃ¬nh
trainer = Trainer(
    model=model,  # Sá»­ dá»¥ng mÃ´ hÃ¬nh Ä‘Ã£ táº¡o trÆ°á»›c Ä‘Ã³
    args=args,  # CÃ¡c tham sá»‘ vÃ  cáº¥u hÃ¬nh huáº¥n luyá»‡n
    train_dataset=train_dataset,  # Sá»­ dá»¥ng táº­p dá»¯ liá»‡u huáº¥n luyá»‡n
    eval_dataset=validation_dataset,  # Sá»­ dá»¥ng táº­p dá»¯ liá»‡u Ä‘Ã¡nh giÃ¡
    tokenizer=tokenizer,  # Sá»­ dá»¥ng tokenizer Ä‘á»ƒ xá»­ lÃ½ vÄƒn báº£n
)

# Báº¯t Ä‘áº§u quÃ¡ trÃ¬nh huáº¥n luyá»‡n
# trainer.train()
```

```python
# Gá»­i dá»¯ liá»‡u Ä‘Ã o táº¡o lÃªn Hub
trainer.push_to_hub(commit_message="Training complete")
```

## **6. Evaluate model**

```python
# Táº£i metric "squad" tá»« thÆ° viá»‡n evaluate
from evaluate import load
metric = load("squad_v2")
```

```python
N_BEST = 20 # Sá»‘ lÆ°á»£ng káº¿t quáº£ tá»‘t nháº¥t Ä‘Æ°á»£c lá»±a chá»n sau khi dá»± Ä‘oÃ¡n
MAX_ANS_LENGTH = 30 # Äá»™ dÃ i tá»‘i Ä‘a cho cÃ¢u tráº£ lá»i dá»± Ä‘oÃ¡n

def compute_metrics(start_logits, end_logits, features, examples):
    # Táº¡o má»™t tá»« Ä‘iá»ƒn máº·c Ä‘á»‹nh Ä‘á»ƒ Ã¡nh xáº¡ má»—i vÃ­ dá»¥ vá»›i danh sÃ¡ch cÃ¡c Ä‘áº·c trÆ°ng tÆ°Æ¡ng á»©ng
    example_to_features = collections.defaultdict(list)
    for idx, feature in enumerate(features):
        example_to_features[feature['example_id']].append(idx)

    predicted_answers = []
    for example in tqdm(examples):
        example_id = example['id']
        context = example['context']
        answers = []

        # Láº·p qua táº¥t cáº£ cÃ¡c Ä‘áº·c trÆ°ng liÃªn quan Ä‘áº¿n vÃ­ dá»¥ Ä‘Ã³
        for feature_index in example_to_features[example_id]:
            start_logit = start_logits[feature_index]
            end_logit = end_logits[feature_index]
            offsets = features[feature_index]['offset_mapping']

            # Láº¥y cÃ¡c chá»‰ sá»‘ cÃ³ giÃ¡ trá»‹ lá»›n nháº¥t cho start vÃ  end logits
            start_indexes = np.argsort(start_logit)[-1 : -N_BEST - 1 : -1].tolist()
            end_indexes = np.argsort(end_logit)[-1 : -N_BEST - 1 : -1].tolist()
            for start_index in start_indexes:
                for end_index in end_indexes:
                    # Bá» qua cÃ¡c cÃ¢u tráº£ lá»i khÃ´ng hoÃ n toÃ n náº±m trong ngá»¯ cáº£nh
                    if offsets[start_index] is None or offsets[end_index] is None:
                        continue
                    # Bá» qua cÃ¡c cÃ¢u tráº£ lá»i cÃ³ Ä‘á»™ dÃ i > max_answer_length
                    if end_index - start_index + 1 > MAX_ANS_LENGTH:
                        continue

                    # Táº¡o má»™t cÃ¢u tráº£ lá»i má»›i
                    answer = {
                        'text': context[offsets[start_index][0] : offsets[end_index][1]],
                        'logit_score': start_logit[start_index] + end_logit[end_index],
                    }
                    answers.append(answer)

        # Chá»n cÃ¢u tráº£ lá»i cÃ³ Ä‘iá»ƒm sá»‘ tá»‘t nháº¥t
        if len(answers) > 0:
            best_answer = max(answers, key=lambda x: x['logit_score'])
            answer_dict = {
                'id': example_id,
                'prediction_text': best_answer['text'],
                'no_answer_probability': 1 - best_answer['logit_score']
            }
        else:
            answer_dict = {
                'id': example_id,
                'prediction_text': '',
                'no_answer_probability': 1.0
            }
        predicted_answers.append(answer_dict)

    # Táº¡o danh sÃ¡ch cÃ¢u tráº£ lá»i lÃ½ thuyáº¿t tá»« cÃ¡c vÃ­ dá»¥
    theoretical_answers = [
        {'id': ex['id'], 'answers': ex['answers']} for ex in examples
    ]
    # Sá»­ dá»¥ng metric.compute Ä‘á»ƒ tÃ­nh toÃ¡n cÃ¡c Ä‘á»™ Ä‘o vÃ  tráº£ vá» káº¿t quáº£
    return metric.compute(predictions=predicted_answers, references=theoretical_answers)
```

```python
# Thá»±c hiá»‡n dá»± Ä‘oÃ¡n trÃªn táº­p dá»¯ liá»‡u validation
predictions, _, _ = trainer.predict(validation_dataset)

# Láº¥y ra thÃ´ng tin vá» cÃ¡c Ä‘iá»ƒm báº¯t Ä‘áº§u vÃ  Ä‘iá»ƒm káº¿t thÃºc cá»§a cÃ¢u tráº£ lá»i dá»± Ä‘oÃ¡n
start_logits, end_logits = predictions

# TÃ­nh toÃ¡n cÃ¡c chá»‰ sá»‘ Ä‘Ã¡nh giÃ¡ sá»­ dá»¥ng hÃ m compute_metrics
results = compute_metrics(
    start_logits,
    end_logits,
    validation_dataset,
    raw_datasets["validation"]
)
results
```
